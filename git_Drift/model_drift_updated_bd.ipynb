### chi square test

Discretization

Prediction probabilities are binned into 10 equal intervals between 0 and 1.

This converts continuous outputs into comparable categorical distributions.

Frequency Normalization

Raw counts are converted into relative frequencies.

Ensures both datasets are comparable even if sizes differ.

Chi-Square Test

Tests whether the current distribution deviates significantly from the reference.

Null hypothesis (H₀): Both distributions are the same.

Decision Rule

p_value < 0.05 → reject H₀ → drift detected

p_value ≥ 0.05 → no statistically significant drift

Define number of bins for probability distribution comparison

```python
n_bins = 10

# Create equally spaced bins between 0 and 1
bins = np.linspace(0, 1, n_bins + 1)
```

```python


# Compute histogram counts for reference (original) probability distribution
ref_counts, _ = np.histogram(orignal_prob, bins=bins)
```

```python
ref_counts
```

```python
# Compute histogram counts for current (sample) probability distribution

cur_counts, _ = np.histogram(sample_prob, bins=bins)
```

```python
cur_counts
```

```python


# Convert histogram counts to relative frequencies for reference data
ref_freq = ref_counts / ref_counts.sum()

# Convert histogram counts to relative frequencies for current data
cur_freq = cur_counts / cur_counts.sum()
```

```python
chi_stat, p_value = chisquare(
    f_obs=cur_freq,
    f_exp=ref_freq
)
```

```python
p_value
```

```python

# Perform Chi-square test to compare current vs reference distributions

if p_value < 0.05:
    drift_status = 'drift'
else:
    drift_status = 'no_drift'
```
